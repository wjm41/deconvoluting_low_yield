This notebook is for calculating RF/GP model metrics to assess their ability to model the underlying dataset of assay measurements

```python
import pandas as pd

gp_df = pd.read_csv('../data/predictions/gp_predictions.csv', index_col='index')
gp_df
```

```python
rf_df = pd.read_csv('../data/predictions/rf_predictions.csv')
rf_df
```

```python
pred_df = rf_df.merge(gp_df, on='Molecule Name')
pred_df['pred'] = pred_df[['RF Predicted Inhibition (%)',
                         'gp_pred']].mean(axis=1) # Average the RF & GP predictions
pred_df
```

Calculate regression metrics first - RMSD, mean abosolute error, and correlations

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
from scipy.stats import spearmanr

def calculate_regression_metrics(df, true_value_col, pred_value_col, label='model'):
    # Extract true values and predicted values from the DataFrame
    y_true = df[true_value_col].values
    y_pred = df[pred_value_col].values

    # Calculate regression metrics
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)  # Root Mean Squared Error
    r2 = r2_score(y_true, y_pred)

    # Calculate Spearman correlation coefficient
    spearman_corr, _ = spearmanr(y_true, y_pred)

    # Create a DataFrame to hold the metrics
    metrics_df = pd.DataFrame({
        'Model': [label],
        'MAE': [mae],
        'RMSE': [rmse],
        'R2': [r2],
        'Spearman_Correlation': [spearman_corr],
    })

    return metrics_df
```

```python
col_model_name_dict = {'pred': 'Mean', 'gp_pred': 'Gaussian Process', 'RF Predicted Inhibition (%)': 'Random Forest'}

regression_metrics = []
for col_name, model in col_model_name_dict.items():
    regression_metrics.append(calculate_regression_metrics(pred_df, 'Inhibition', col_name, label=model))

regression_metrics = pd.concat(regression_metrics).reset_index(drop=True)
regression_metrics
```

What if we treat as binary classification? Label dataset by inhibition value, and use the models as classifiers. Relevant classification metrics are accuracy, precision, matthews correlation correction, as well as ROC-AUC and precision-recall curve AUC (PRC-AUC).

```python
import pandas as pd
from sklearn.metrics import roc_auc_score, matthews_corrcoef, confusion_matrix, accuracy_score, precision_score, auc, precision_recall_curve

def calculate_metrics(df, true_label_col, pred_score_col, threshold=50, label=None):
    if label is None:
        label = 'model'

    # True binary labels
    y_true = df[true_label_col].values > threshold

    # Binary predictions based on the threshold
    y_pred = df[pred_score_col].values > threshold

    # Prediction scores for PRC-AUC calculation
    y_scores = df[pred_score_col].values

    # Calculate metrics
    mcc = matthews_corrcoef(y_true, y_pred)
    cm = confusion_matrix(y_true, y_pred).ravel()
    tn, fp, fn, tp = cm if len(cm) == 4 else (0, 0, 0, 0)  # Handle case with less than 4 values
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)
    precision_points, recall_points, _ = precision_recall_curve(y_true, y_scores)
    prc_auc = auc(recall_points, precision_points)
    roc_auc = roc_auc_score(y_true, y_pred)


    # Create a DataFrame to hold the metrics
    metrics_df = pd.DataFrame({
        'Model': [label],
        'MCC': [mcc],
        'Accuracy': [accuracy],
        'Precision': [precision],
        'ROC-AUC': [roc_auc],
        'PRC-AUC': [prc_auc],
        'TP': [tp],
        'FP': [fp],
        'FN': [fn],
        'TN': [tn],
    })

    return metrics_df
```

```python
col_model_name_dict = {'pred': 'Mean', 'gp_pred': 'Gaussian Process', 'RF Predicted Inhibition (%)': 'Random Forest'}

classification_metrics = []
for col_name, model in col_model_name_dict.items():
    classification_metrics.append(calculate_metrics(pred_df, 'Inhibition', col_name, label=model, threshold=70))

classification_metrics = pd.concat(classification_metrics).reset_index(drop=True)
classification_metrics
```

Here we see the power of using the mean of the two models in terms of precision/recall!

```python
with pd.ExcelWriter('model_metrics.xlsx') as writer:
    regression_metrics.to_excel(writer, sheet_name='Regression Metrics', index=False)
    classification_metrics.to_excel(writer, sheet_name='Classification Metrics', index=False)
```
