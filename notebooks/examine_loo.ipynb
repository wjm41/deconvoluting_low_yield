```python
import pandas as pd

inliers = pd.read_csv('../data/predictions/intersected_inliers.csv', index_col='index')
inliers
```

```python
outliers = pd.read_csv('../data/predictions/intersected_outliers.csv', index_col='index')
outliers
```

```python
gp_df = pd.concat([inliers, outliers]).sort_index()
gp_df['gp_pred'] = gp_df[[
    'amine_rbf_pred', 'amine_matern_pred']].mean(axis=1)
gp_df
```

```python
rf_df = pd.read_csv('../data/predictions/loo_predictions.csv')
rf_df
```

```python
loo_df = rf_df.merge(gp_df, on='Molecule Name')
loo_df['pred'] = loo_df[['RF Predicted Inhibition (%)',
                         'gp_pred']].mean(axis=1) # Average my models and emmas models
loo_df
```

Treat as binary classification

```python
import pandas as pd
from sklearn.metrics import matthews_corrcoef, confusion_matrix, accuracy_score, precision_score, auc, precision_recall_curve

def calculate_metrics(df, true_label_col, pred_score_col, threshold=50, label=None):
    if label is None:
        label = 'model'

    # True binary labels
    y_true = df[true_label_col].values > threshold

    # Binary predictions based on the threshold
    y_pred = df[pred_score_col].values > threshold

    # Prediction scores for PRC-AUC calculation
    y_scores = df[pred_score_col].values

    # Calculate metrics
    mcc = matthews_corrcoef(y_true, y_pred)
    cm = confusion_matrix(y_true, y_pred).ravel()
    tn, fp, fn, tp = cm if len(cm) == 4 else (0, 0, 0, 0)  # Handle case with less than 4 values
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)
    precision_points, recall_points, _ = precision_recall_curve(y_true, y_scores)
    prc_auc = auc(recall_points, precision_points)

    # Create a DataFrame to hold the metrics
    metrics_df = pd.DataFrame({
        'Model': [label],
        'MCC': [mcc],
        'Accuracy': [accuracy],
        'Precision': [precision],
        'PRC-AUC': [prc_auc],
        'TP': [tp],
        'FP': [fp],
        'FN': [fn],
        'TN': [tn],
    })

    return metrics_df

# Example usage
# Assuming you have a DataFrame named `loo_df` with columns 'Inhibition' for true labels
# and 'pred' for prediction scores
metrics_df = calculate_metrics(loo_df, 'Inhibition', 'pred', label='Mean', threshold=70)
print(metrics_df)
```

```python
col_model_name_dict = {'pred': 'Mean', 'gp_pred': 'Gaussian Process', 'RF Predicted Inhibition (%)': 'Random Forest'}

classification_metrics = []
for col_name, model in col_model_name_dict.items():
    classification_metrics.append(calculate_metrics(loo_df, 'Inhibition', col_name, label=model, threshold=70))

classification_metrics = pd.concat(classification_metrics).reset_index(drop=True)
classification_metrics
```

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
from scipy.stats import spearmanr

def calculate_regression_metrics(df, true_value_col, pred_value_col, label='model'):
    # Extract true values and predicted values from the DataFrame
    y_true = df[true_value_col].values
    y_pred = df[pred_value_col].values

    # Calculate regression metrics
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)  # Root Mean Squared Error
    r2 = r2_score(y_true, y_pred)

    # Calculate Spearman correlation coefficient
    spearman_corr, _ = spearmanr(y_true, y_pred)

    # Create a DataFrame to hold the metrics
    metrics_df = pd.DataFrame({
        'Model': [label],
        'MAE': [mae],
        'RMSE': [rmse],
        'R2': [r2],
        'Spearman_Correlation': [spearman_corr],
    })

    return metrics_df

# Example usage
# Assuming you have a DataFrame named `regression_df` with columns 'Actual' for true values
# and 'Predicted' for predicted values
metrics_df = calculate_regression_metrics(loo_df, 'Inhibition', 'pred')
print(metrics_df)
```

```python
col_model_name_dict = {'pred': 'Mean', 'gp_pred': 'Gaussian Process', 'RF Predicted Inhibition (%)': 'Random Forest'}

regression_metrics = []
for col_name, model in col_model_name_dict.items():
    regression_metrics.append(calculate_regression_metrics(loo_df, 'Inhibition', col_name, label=model))

regression_metrics = pd.concat(regression_metrics).reset_index(drop=True)
regression_metrics
```

```python
with pd.ExcelWriter('model_metrics.xlsx') as writer:
    regression_metrics.to_excel(writer, sheet_name='Regression Metrics', index=False)
    classification_metrics.to_excel(writer, sheet_name='Classification Metrics', index=False)
```
